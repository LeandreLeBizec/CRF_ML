```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def load(file):
    with open(file,'r') as fichier:
        X=[]
        Y=[]
        phraseX=[]
        phraseY=[]
        for line in fichier:
            if(len(line)==2):
                X.append(phraseX)
                Y.append(phraseY)
                phraseX=[]
                phraseY=[]
            else:
                mot = line.split("\t")
                phraseX.append(mot[0])
                phraseY.append(mot[1].replace("\n",""))
        return X, Y


Sentences, PostTags = load("Data/atis.train")
SentencesTrain, SentencesTest, PostTagsTrain, PostTagsTest = train_test_split(Sentences, PostTags, test_size=0.5, random_state=0)
```

```python
def USCities(): 
    USCities = []
    US_Cities = pd.read_csv("Data/uscities.csv")
    US_Cities_name = US_Cities['city']
    for cities in US_Cities_name:
        #USCities.append(cities.lower())
        cities = cities.replace(" ", "-")
        USCities.append(cities.lower())
    return USCities

USCities = USCities()
USCities
```

```python
Days = ['monday', 'tuesday', 'wednesday', 'thursday', 'Friday', 'saturday', 'sunday']
```

```python
def cityTag(Sentences):
    SentencesTag = []
    for sentence in Sentences:
        SentenceTag = []
        for word in sentence:
            if(word in USCities):
                WordTag = 'city'
            elif (word in Days):
                WordTag = 'day'
            else:
                WordTag = 'O'
            SentenceTag.append(WordTag)
        SentencesTag.append(SentenceTag)
    return SentencesTag
    
TagsTrain = cityTag(SentencesTrain)
TagsTest = cityTag(SentencesTest)
```

```python
def write(filename, Sentences, PostTags, Tags):
    # Ouvrir un fichier texte en √©criture
    with open("Data/" + filename, "w") as f:
        # Parcourir les tableaux et √©crire chaque ligne dans le fichier
        for i in range(len(Sentences)):
            for j in range(len(Sentences[i])):
                ligne = f"{Sentences[i][j]} {Tags[i][j]} {PostTags[i][j]}\n"
                f.write(ligne)
            f.write("\n")

write("atis.tain", SentencesTrain, PostTagsTrain, TagsTrain)
write("atis.test", SentencesTest, PostTagsTest, TagsTest)
```
